# Cau 1

```python
from sklearn.metrics import mean_squared_error
z = {(1,1), (2,3), (3, 2), (4,4), (5, 4)}

x=[]
y_t=[]

for i,j in z:
    x.append(i)
    y_t.append(j)

#a
y_p = [2*i+1 for i in x]

def MSE(y,y_hat):
    x=0
    for i in range(len(y)):
        x += (y[i]-y_hat[i])**2    
    x = 1/len(y)*x
    return float("{:.2f}".format(x))

print(MSE(y_p,y_t))


#b
def cal_MSE():
    a = float(input('nhap a: '))
    b = float(input('nhap b: '))
    y_p = [a*i+b for i in x]
    return MSE(y_p,y_t)

# print(cal_MSE())

#c
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error

X=np.array(x).reshape(-1,1)
y=np.array(y_t).reshape(-1,1)

reg = LinearRegression().fit(X, y)
y_p = reg.predict(np.array(X))
print('MSE',mean_squared_error(y, y_p))
print('coef',reg.coef_)
print('intercept',reg.intercept_)

import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(X.reshape(-1),y.reshape(-1),color='green',markers='+')
sns.lineplot(X.reshape(-1), y_p.reshape(-1), color="pink", linewidth=3)
```

# Cau 2

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('data/data_linear.csv')

# plt.scatter(df['Diện tích'], df['Giá'], color="black")
sns.scatterplot(df['Diện tích'], df['Giá'],color='brown')
plt.xlabel('Diện tích')
plt.ylabel('Giá')
plt.show()
```

```python
X = df['Diện tích'].to_numpy().reshape(-1,1)
y = df['Giá'].to_numpy().reshape(-1,1)
reg = LinearRegression().fit(X, y)
y_p = reg.predict(X)
print('MSE',mean_squared_error(y, y_p))
print('coef',reg.coef_)
print('intercept',reg.intercept_)
```

```python
sns.scatterplot(df['Diện tích'], df['Giá'],color='green')
sns.lineplot(df['Diện tích'], y_p.reshape(-1), color="pink", linewidth=3)
plt.xlabel('Diện tích')
plt.ylabel('Giá')
plt.title('Cau 2')
plt.show()
```

# Cau 3

```python
df = pd.read_csv('data/data_square.csv')

plt.scatter(df['Diện tích'], df['Giá nhà'], color="black")
plt.xlabel('Diện tích')
plt.ylabel('Giá')
plt.show()
```

```python
X = df['Diện tích'].to_numpy().reshape(-1,1)
y = df['Giá nhà'].to_numpy().reshape(-1,1)
reg = LinearRegression().fit(X, y)
y_p = reg.predict(X)
print('MSE',mean_squared_error(y, y_p))
print('coef',reg.coef_)
print('intercept',reg.intercept_)
```

```python
plt.scatter(X, y, color="green")
plt.plot(X, y_p, color="pink", linewidth=3)
plt.xlabel('Diện tích')
plt.ylabel('Giá')
plt.title('Underfitting')
plt.show()
```
