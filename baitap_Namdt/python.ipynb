```python
num = [1, 2, 3, 1, 2, 1, 1,4] 

def CntFreq(list):
    dic = {}
    for i in list:
        if (i in dic):
            dic[i] += 1
        else:
            dic[i] = 1
    return dic
CntFreq(num)
```

```python
def MaxMinfreq(dict):
    ma=-69
    ma_list=[]
    mi= 69
    mi_list=[]
    for v in dict.values():
        if ma<v: ma =v
            
        if mi>v: mi =v

    for k,v in dict.items():
        if ma == v: ma_list.append(k)
        if mi == v: mi_list.append(k)

    print('max', ma_list)
    print('min', mi_list)

MaxMinfreq(dic)
```

# Cau 14


```python
import string
def v_ing(stri):
    stri = stri.lower()
    letter = list(string.ascii_lowercase)
    vowel = ['a','e','i','o','u']
    consonant = [c for c in letter if c not in vowel]
    
    if stri.endswith('ie'):
        stri = stri[:-2]
        return stri + 'ying'

    elif stri.endswith('e'):
        stri = stri[:-1]
        return stri + 'ing'

    elif stri[-1] in consonant and stri[-2] in vowel and stri[-3] in consonant:
        stri += stri[-1]
        return stri + 'ing'
    else:
        return stri + 'ing'

verb = ['move', 'hug', 'lie']
for item in verb:
    print(v_ing(item))
```

# Cau 16


```python
def replace():
    letter = input()
    word = input()
    reword = input()
    pos = int(input())
    
    if pos == '':
        letter = letter.split()
        for i in range(len(letter)):
            if letter[i]==word: letter[i]=reword

    if pos != '':
        cnt=0
        letter = letter.split()
        for i in range(len(letter)):
            if letter[i]==word: 
                cnt+=1
                if cnt-1 == pos: 
                    letter[i]=reword  
    
    return ' '.join(letter)        
        
replace()
```
