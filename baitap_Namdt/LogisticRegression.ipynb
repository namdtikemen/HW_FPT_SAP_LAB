```python
import math
def sigmoid(x):
    a=[]
    if type(x)==list:
        for i in x:
            a.append(1 / (1 + math.exp(-i)))
        return a
    else: return 1 / (1 + math.exp(-x))
sigmoid([-4, -2, -1, 0, 1, 3, 6])
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = [i for i in np.arange(-10., 10., 0.2)]
y = sigmoid(x)
plt.plot(x,y)
plt.show()
```

```python

y = [1., 0., 1., 1., 0.]
y_predict = [0.8, 0.5, 0.9, 0.9, 0.1]
y = np.array(y)
y_predict = np.array(y_predict)

def binary_crossentropy(y,y_p):
    a = y*np.log(y_p) + (1-y)*np.log(1-y_p)
    a = -1/(len(y))*np.sum(a)
    return a

binary_crossentropy(y,y_predict)
```

```python
import sklearn.metrics
sklearn.metrics.log_loss(y, y_predict)
```

```python
import matplotlib.pyplot as plt

# Import datasets, classifiers and performance metrics
from sklearn import datasets, metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
digits = datasets.load_digits()
n_samples = len(digits.images)
data = digits.images.reshape((n_samples, -1))
X_train, X_test, y_train, y_test = train_test_split(
    data, digits.target, test_size=0.3, shuffle=False
)

clf = LogisticRegression()
clf.fit(X_train, y_train)
predicted = clf.predict(X_test)

print(metrics.classification_report(y_test, predicted))
```

```python
visual = metrics.ConfusionMatrixDisplay.from_predictions(y_test, predicted)
visual.figure_.suptitle("Confusion Matrix")
print(f"Confusion matrix:\n{visual.confusion_matrix}")

plt.show()
```
